```python
import seaborn as sns 
import pandas as pd 
```

```python
sales= pd.read_excel('sales.xlsx')
pd.set_option('display.max_columns',25)
```

```python
sales.head()
```

```python
sns.catplot(x='Product Category',y='Sales',data=sales,kind='point')
```

```python
sns.catplot(x='Product Category',y='Sales',hue = 'Ship Mode',data=sales,kind='point')
```

```python
sns.catplot(x='Product Category',y='Sales',hue='Region', row = 'Region',data=sales,kind='point',ci=None)
```
