```python
import seaborn as sns 
import pandas as pd 
import numpy as np
```

```python
sales= pd.read_excel('sales.xlsx')
```

```python
sales.head()
```

```python
sns.jointplot('Order Quantity','Sales',data=sales)
```

```python
sns.jointplot('Sales','Discount',data=sales)
```

```python
sns.jointplot('Ship Mode','Sales',data=sales)
```

```python
sns.jointplot('Sales','Product Base Margin',data=sales,kind='hex')
```

```python
sns.jointplot('Discount','Product Base Margin',data=sales,kind='hex')
```

```python
sns.jointplot("Discount","Product Base Margin",data=sales,kind='kde', ci=True)
```

```python
sns.pairplot(sales,hue='Ship Mode')
```
