```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
sales= pd.read_excel('sales.xlsx')
```

```python
sales.head()
```

```python
a=sns.relplot('Order Date','Sales',data=sales,kind='line')
```

```python
a
a.figure.savefig('output1.png')
```

```python
sales['year']=sales ['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
b=sns.relplot('year','Sales',data=sales,kind='line')
```

```python
b.figure.savefig('output2.png')
```

```python
c=sns.relplot('year','Sales',data=sales,kind='line',ci=None)
c.figure.savefig('output3.png')
```

```python
d=sns.relplot('year','Sales',data=sales,kind='line',estimator=None)
d.figure.savefig('output4.png')
```

```python
e=sns.relplot('year','Sales',data=sales,kind='line',estimator=sum,ci=None)
e.figure.savefig('output5.png')
```

```python
f=sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode')
f.figure.savefig('output6.png')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Product Category',style='Ship Mode',ci=None,col='Ship Mode')
```

```python
g=sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category',ci=None,col='Region',col_wrap=4)
g.figure.savefig('output7.png')
```
