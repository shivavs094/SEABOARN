```python
import seaborn as sns 
import pandas as pd 
```

```python
sales= pd.read_excel('sales.xlsx')
```

```python
sales.head(4)
```

```python
a=sns.relplot('Sales','Profit',data = sales,col='Customer Segment',row='Product Category',hue='Ship Mode')
a.figure.savefig('output1.png')
```

```python
b=sns.relplot('Sales','Profit',data = sales,col='Customer Segment',hue='Ship Mode',col_wrap=2)
b.figure.savefig('output5.png')
```
