```python
import seaborn as sns 
import pandas as pd 
```

```python
sales= pd.read_excel('sales.xlsx')
pd.set_option('display.max_column',25)
```

```python
sales.head ()
```

```python
a=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,kind='box')
a.figure.savefig('Box_1.png')
```

```python
b=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,hue='Ship Mode',kind='box')
b.figure.savefig('Box_2.png')
```

```python
c=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,hue='Ship Mode',hue_order=['Delivery Truck','Regulr Air','Express Air'],kind='box')
c.figure.savefig('Box_3.png')
```

```python
d=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,hue='Ship Mode',col="Ship Mode",kind='box')
d.figure.savefig('Box_4.png')
```

```python
e=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,col="Ship Mode",kind='box')
e.figure.savefig('Box_5.png')
```

```python
f=sns.catplot(x="Product Category",y='Shipping Cost',data=sales,hue='Ship Mode',col="Ship Mode",row='Customer Segment' ,kind='box')
f.figure.savefig('Box_6.png')
```
